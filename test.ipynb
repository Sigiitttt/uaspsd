```python
import numpy as np
import random # Tambahkan library random

# --- Script Pembuat File Dummy (Versi RANDOM & Encoding Fix) ---
source_path = "dataset/SemgHandSubjectCh2_TEST.ts"
output_path = "contoh_sinyal_untuk_upload.txt"

print("üé≤ Sedang mengambil sampel ACAK dari dataset...")

# 1. Baca File dengan Encoding yang Aman
encodings = ['utf-8', 'latin-1'] 
lines = []
read_success = False

for enc in encodings:
    try:
        with open(source_path, 'r', encoding=enc) as f:
            lines = f.readlines()
        read_success = True
        break
    except UnicodeDecodeError:
        continue

if not read_success:
    print("‚ùå Gagal membaca file sumber.")
else:
    try:
        # 2. Filter hanya baris data (buang metadata @...)
        data_lines = [line for line in lines if not line.startswith(('@', '%', '#')) and line.strip()]
        
        if len(data_lines) > 0:
            # --- BAGIAN INI YANG DIUBAH JADI RANDOM ---
            sample_line = random.choice(data_lines) 
            # ------------------------------------------
            
            # 3. Pisahkan Data & Label
            if ':' in sample_line:
                parts = sample_line.split(':')
                raw_values_str = parts[0]
                true_label = parts[1].strip()
            else:
                parts = sample_line.split(',')
                raw_values_str = ",".join(parts[:-1])
                true_label = parts[-1].strip()

            # 4. Simpan ke File TXT (Menimpa file lama)
            with open(output_path, "w", encoding='utf-8') as f:
                f.write(raw_values_str)

            print("\n BERHASIL MEMBUAT FILE BARU!")
            print(f"Nama File : {output_path}")
            print(f"Kunci Jawaban (Label Asli): SUBJEK {true_label}")
            print("-" * 40)
            print("Sekarang upload file ini ke Streamlit.")
            print("Cek apakah Prediksi AI cocok dengan Kunci Jawaban di atas.")
            
        else:
            print("‚ùå File kosong.")

    except Exception as e:
        print(f"‚ùå Error: {e}")
```
